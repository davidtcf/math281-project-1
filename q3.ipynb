$$\LARGE\text{ODE \#1:}$$
$$\hspace{0.75cm}$$
$$\frac{du}{dt}=\sin(2\pi t)\implies$$
$$u(t)=-\frac{1}{2\pi}(\cos(2\pi t)-1)$$

$$\underline{\Delta t = 0.01}$$

```python
import math
import numpy as np
import matplotlib.pyplot as plt

u=0
t=0
tf=2
deltat=0.01

ulist=[]
tlist=[]
solutionlist=[]
errorlist=[]

while t<tf+deltat:
    dudt=math.sin(2*math.pi*t)
    u=u+deltat*dudt

    solution=(-(1/(2*math.pi)))*(math.cos(2*math.pi*t)-1)

    ulist.append(u)
    tlist.append(t)
    solutionlist.append(solution)

    t=t+deltat

uarray=np.array(ulist)
tarray=np.array(tlist)
solutionarray=np.array(solutionlist)

gridspec_props = dict(wspace = 0.05, hspace = 0.3, left = 0.1, right = 0.75, bottom = 0.1, top = 0.9)

fig, axes = plt.subplots(2, 1, sharex=False, figsize = (10, 8), gridspec_kw = gridspec_props)

axes[0].plot(tarray, uarray,          label = 'Approximated\nu(t), '+r'$\Delta$t'+f'= {deltat}')
axes[0].plot(tarray, solutionarray, '--', label = 'True solution')
axes[1].plot(tarray, uarray - solutionarray, label = 'Error in\nApproximated\nu(t)')

axes[0].legend(bbox_to_anchor=(1., 0, 0.25, 1))
axes[1].legend(bbox_to_anchor=(1., 0, 0.25, 1))

axes[0].set_xlabel('Time')
axes[1].set_xlabel('Time')
axes[0].set_ylabel('u(t)')
axes[1].set_ylabel('Approximation Error')

axes[0].set_title(r"Approximated u(t) and Exact u(t) for $\Delta t$=0.01")
axes[1].set_title(r"Evolution in Error of Approximated u(t) for $\Delta t$=0.01")

axes[0].set_xticks(np.arange(0, 2.25, 0.25))
axes[0].set_xticklabels(np.arange(0, 2.25, 0.25))
axes[1].set_xticks(np.arange(0, 2.25, 0.25))
axes[1].set_xticklabels(np.arange(0, 2.25, 0.25))

plt.savefig("ODE1dt001.png")
plt.show()
```

$$\LARGE\text{ODE \#2:}$$
$$\hspace{0.75cm}$$
$$\frac{du}{dt}=\frac{u}{2}\implies$$
$$u(t)=e^{t/2}$$

$$\underline{\Delta t=0.001}$$

```python
u=0.01
t=0
tf=100
deltat=0.001

ulog=[]
tlist=[]
logsolution=[]
differencelist=[]

while t<tf+deltat:
    dudt=(math.e**(t/2))/2
    u=u+deltat*dudt

    solution=math.e**(t/2)-0.99

    ulog.append(math.log(u))
    tlist.append(t)
    logsolution.append(math.log(solution))
    differencelist.append(math.log(abs(u-solution)))

    t=t+deltat

uarray=np.array(ulist)
tarray=np.array(tlist)
ulogarray=np.array(ulog)
solutionarray=np.array(solutionlist)
logsolarray=np.array(logsolution)
differencearray=np.array(differencelist)

gridspec_props = dict(wspace = 0.05, hspace = 0.3, left = 0.1, right = 0.75, bottom = 0.1, top = 0.9)

fig, axes = plt.subplots(2, 1, sharex=False, figsize = (10, 8), gridspec_kw = gridspec_props)


axes[0].plot(tarray, ulogarray, label = 'ln of\nApproximated\nu(t), '+r'$\Delta$t'+f'= {deltat}' )
axes[0].plot(tarray,logsolarray, '--', label = 'ln of True\nSolution')
axes[1].plot(tarray, differencearray, label = 'Error in\nApproximated\nu(t)')


axes[0].legend(bbox_to_anchor=(1.01,0,0.25,1))
axes[1].legend(bbox_to_anchor=(1., 0, 0.25, 1))


axes[0].set_xlabel('Time')
axes[1].set_xlabel('Time')
axes[0].set_ylabel('ln(u(t))')
axes[1].set_ylabel('ln|Approximation Error|')

axes[0].set_title(r"ln of Approximated u(t) and ln of Exact u(t) for $\Delta t$=0.001")
axes[1].set_title(r"Evolution in ln of the Error in Approximated u(t) for $\Delta t$=0.001")

axes[0].set_xticks(np.arange(0, 105, 10))
axes[0].set_xticklabels(np.arange(0, 105, 10))
axes[1].set_xticks(np.arange(0, 105, 10))
axes[1].set_xticklabels(np.arange(0, 105, 10))

plt.savefig("ODE2dt0001.png")
plt.show()
```
